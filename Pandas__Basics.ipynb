# Pandas Basics

```python
import numpy as np
import pandas as pd
```

## Series

```python
list1=[10,20,30]
series=pd.Series(data=list1)
series
```

```python
arr1=np.arange(21,31)
arr1
```

```python
series1=pd.Series(data=arr1,index=['a','b','c','d','e','f','g','h','i','j'])
series1
```

```python
series1[3]
```

```python
series2=pd.Series([1,2,3,4,5],['a','b','c','d','e'])
series2
```

```python
series3=pd.Series([10,20,30,40,50],['f','g','h','i','j'])
series3
```

## Operations on Series

```python
series2+series3 #If index doesn't match addition operation cannot be performed
```

```python
series4=pd.Series([11,22,33,44,55],['a','b','c','d','e'])
series4
```

```python
series2+series4 #Index of both the Series are same so adding is possible
```

```python
series5=pd.Series([12,22,23,24,35],['a','c','d','f','e'])
series5
```

```python
series4+series5
```

## Creating a DataFrame with Missing Values

```python
data1={'A':[np.nan,1,3],'B':[5,7,np.nan],'C':[np.nan,9,np.nan]}
df2=pd.DataFrame(data=data1)
df2
```

```python
df2['A']
```

```python
df2.fillna(value=0)
```

```python
df2['B'].fillna(value=50)
```

```python
df2.iloc[0]
```

```python
df2.sum()
```
